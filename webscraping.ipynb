```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_companies_of_Nepal'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[0]
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
all_ths = table.find_all('th')
```

```python
all_ths
```

```python
filtered_ths = all_ths[:7]
```

```python
filtered_ths
```

```python
company_table_title = [title.text.strip() for title in filtered_ths]
```

```python
company_table_title
```

```python
df = pd.DataFrame(columns = company_table_title)
```

```python
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[2:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    new_individual_data = individual_row_data[:7]
    length = len(df)
    df.loc[length] = new_individual_data
```
